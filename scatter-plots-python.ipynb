## Load dependencies and Dataset

```python
# load dependencies
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
import plotly.express as px
from bokeh.plotting import figure, show
```

```python
# load the dataset
filename = "iris.data"

# column names to use
names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']

# read the dataset from URL
dataset = pd.read_csv(filename, names=names) 

# check first few rows of data
dataset.head()
```

## Scatter Plot Matplotlib

```python
# configure plot
plt.title('Scatter Plot with Matplotlib')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# create scatter plot 
plt.scatter(dataset['sepal-length'], dataset['sepal-width'])

# show plot 
plt.show()
```

## Scatter plot Seaborn

```python
# configure plot
plt.title('Scatter Plot with Seaborn')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# create scatter plot with seaborn
sns.scatterplot(x = 'sepal-length', y = 'sepal-width', hue = 'Class', data=dataset)

# show plot 
plt.show()
```

## Scatter Plot Plotly

```python
# create scatter plot with plotly
fig = px.scatter(dataset, x = 'sepal-length', y = 'sepal-width', 
                 color = 'Class', symbol = 'Class', title = 'Scatter Plot with Plotly')

# show plot
fig.show()
```

## Scatter Plot Bokeh

```python
# create a new plot
p = figure(title = "Scatter Plot with Bokeh")

# scatter plot with bokeh
p.scatter('sepal-length','sepal-width',source = dataset, fill_alpha = 0.2, size = 10)
p.xaxis.axis_label = 'sepal-length'
p.yaxis.axis_label = 'sepal-width'
p.legend.location = "top_left"

# show plot
show(p)
```
